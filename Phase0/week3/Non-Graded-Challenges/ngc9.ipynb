### Scrape coins price from https://coinmarketcap.com/ with following requirements:
##### ● Scrape up to 1000 coins
##### ● Including Price, Market Cap, Volume, and Coin supply
##### ● Save into Pandas DataFrame and then save into excel or csv file4


```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
from bs4 import BeautifulSoup
driver = webdriver.Chrome()

names = []
prices = []
market_caps = []
volumes = []
circulating_supplies = []
for index in range(10):
  url = f"https://coinmarketcap.com/?page={index + 1}"
  driver.get(url)
  html = driver.page_source

  page = BeautifulSoup(html, "html.parser")

  result = page.find_all("tr", {"style": "cursor:pointer"})
  for i in result:
    paragraph = i.find("p", {"font-weight": "semibold"})
    if (paragraph):
      names.append(paragraph.get_text())
    else:
      names.append(None)
    div_price = i.find("div", {"class": "sc-b3fc6b7-0 dzgUIj"})
    if div_price:
      span = div_price.find("span")
      if span:
        prices.append(span.get_text())
      else:
        prices.append(None)
    else:
      prices.append(None)
    span_marketcap = i.find("span", {"class": "sc-11478e5d-1 jfwGHx"})
    if span_marketcap:
      market_caps.append(span_marketcap.get_text())
    else:
      market_caps.append(None)
    p_volume = i.find("p", {"class": "sc-71024e3e-0 bbHOdE font_weight_500"})
    if p_volume:
      volumes.append(p_volume.get_text())
    else:
      volumes.append(None)
    p_circulating_supply = i.find("p", {"class": "sc-71024e3e-0 hhmVNu", "color": "text", "font-size": "1"})
    if p_circulating_supply:
      circulating_supplies.append(p_circulating_supply.get_text())
    else:
      circulating_supplies.append(None)

```

```python
import pandas as pd
```

```python
df = pd.DataFrame()

df["name"] = names
df["price"] = prices
df["market_cap"] = market_caps
df["volume"] = volumes
df["circulating_supply"] = circulating_supplies
df.to_csv("coinmarketcap.csv", index=False)
df
```

### Get weather data from 54 locations in Indonesia from https://www.timeanddate.com/weather/indonesia. 
### You should include the information of:
##### ● Temperature
##### ● Feels Like
##### ● Wind Speed
##### ● Wind Direction
##### ● Humidity
##### ● Dew Point
##### ● Visibility
##### ● Probability of Rain
##### ● Amount of Rain
### Only take the closest day and time (morning/evening) from the time you do this task
### Save into Pandas DataFrame and then save into excel or csv file5


```python
from bs4 import BeautifulSoup
driver = webdriver.Chrome()

url = "https://www.timeanddate.com/weather/indonesia"
driver.get(url)
html = driver.page_source
page = BeautifulSoup(html, "html.parser")

result = page.find_all("tr")
# print(result)
cities = []
for i in result:
  anchors = i.find_all("a")
  if anchors:
    for anchor in anchors:
      if anchor:
        cities.append(anchor.get_text())

print(cities)
location = []
temperature = []
feels_like = []
wind_speed = []
wind_direction = []
humidity = []
dew_point = []
visibility = []
probability_of_rain = []
amount_of_rain = []

def append_all_none():
  location.append(None)
  temperature.append(None)
  feels_like.append(None)
  wind_speed.append(None)
  wind_direction.append(None)
  humidity.append(None)
  dew_point.append(None)
  visibility.append(None)
  probability_of_rain.append(None)
  amount_of_rain.append(None)

for j in cities:
  url = f"https://www.timeanddate.com/weather/indonesia/{j.lower()}"
  driver.get(url)
  html = driver.page_source
  page = BeautifulSoup(html, "html.parser")

  location.append(j)
  result = page.find("div", {"class": "eight columns"})
  if result:
    table = result.find("table")
    print(table)
    if table:
      rows = table.find_all("tr")
      if rows:
        for k in rows:
          thead = k.find("th")
          td = k.find("td")
          if thead and td:
            if thead.get_text().lower() == 'temperature':
              temperature.append(td.get_text())
            if thead.get_text().lower() == 'feels like':
              feels_like.append(td.get_text())
            if thead.get_text().lower() == 'wind speed':
              wind_speed.append(td.get_text())
            if thead.get_text().lower() == 'humidity':
              humidity.append(td.get_text())
            if thead.get_text().lower() == 'dew point':
              dew_point.append(td.get_text())
            if thead.get_text().lower() == 'visibility':
              visibility.append(td.get_text())
            if thead.get_text().lower() == 'probability of rain':
              probability_of_rain.append(td.get_text())
            if thead.get_text().lower() == 'amount of rain':
              amount_of_rain.append(td.get_text())
          else:
            append_all_none()
      else:
        append_all_none()
    else:
      append_all_none()
  else:
    append_all_none()

print(temperature)
print(feels_like)
print(wind_speed)
print(humidity)
print(dew_point)
print(visibility)
print(probability_of_rain)
print(amount_of_rain)
        
```
