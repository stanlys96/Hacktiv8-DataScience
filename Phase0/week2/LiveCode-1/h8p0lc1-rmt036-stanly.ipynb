Program ini dibuat untuk melakukan perhitungan penghasilan karyawan berdasarkan posisi dan total jam kerja mereka.
=================================================

```python
# dictionary posisi karyawan dan gajinya
penghasilan_data = {
  "manager": 10000000,
  "staff": 5000000,
  "kasir": 4000000
}

def hitung_penghasilan(position, work_hours) -> int:
  '''
  Fungsi ini ditujukan untuk menghitung jumlah total penghasilan 1 karyawan berdasarkan jumlah total kerja
  yang telah dilakukan

  Argument positions merupakan inputan string, sedangkan argument work_hours merupakan inputan integer

  Jika input jenis pekerjaan tidak ada di dalam key dictionary, maka fungsi ini akan me-return nilai 0

  Contoh penggunaan:
    result = hitung_penghasilan("manager", 45)
    print(result)
    --------
    Output: 10100000
  '''
  overtime_pay = (work_hours - 40) * 20000 if work_hours > 40 else 0
  try:
    return penghasilan_data[position.lower()] + overtime_pay
  except:
    return 0

print(hitung_penghasilan("manager", 45))

def total_pengeluaran(data_karyawan) -> int:
  '''
  Fungsi ini ditujukan untuk menghitung jumlah total pengeluaran yang harus dilakukan olah perusahaan. Inputnya
  bisa lebih dari 1 karyawan

  Argument data_karyawan merupakan sebuah dictionary, dengan key sebagai jenis pekerjaan, dan value sebagai total jam kerja
  yang telah dilakukan oleh karyawan tersebut

  Jika input jenis pekerjaan tidak ada di dalam key dictionary, maka jenis pekerjaan tersebut akan dianggap pengeluarannya 0

  Contoh penggunaan:
    karyawan_data = {
      "manager": 50,
      "staff": 50,
      "kasir": 50
    }

    result = total_pengeluaran(karyawan_data)
    print(result)
    --------
    Output: 19600000
  '''
  total = 0
  for key, value in data_karyawan.items():
    try:
      total += hitung_penghasilan(key, value)
    except:
      pass

  return total

karyawan_data = {
  "manager": 50,
  "staff": 50,
  "kasir": 50,
  "pilot": 100
}

karyawan_data_2 = {
  "manager": 45,
  "staff": 35,
  "kasir": 50,
}

print(total_pengeluaran(karyawan_data))
print(total_pengeluaran(karyawan_data_2))
```

```python
import unittest
from unittest.case import _AssertRaisesContext

class PenghasilanTest(unittest.TestCase):
  def test_hitung_penghasilan(self):
    '''
    Fungsi ini ditujukan untuk testing apakah fungsi hitung_penghasilan berjalan sesuai ekspektasi
    '''
    self.assertEqual(hitung_penghasilan("manager", 45), 10100000)

  def test_hitung_penghasilan_jenis_pekerjaan_tidak_ada(self):
    '''
    Fungsi ini ditujukan untuk testing apakah fungsi hitung_penghasilan akan me-return nilai 0 jika data "position"
    tidak ada di dalam list keys dictionary
    '''
    self.assertEqual(hitung_penghasilan("pilot", 100), 0)

  def test_hitung_penghasilan_data_type_salah(self) -> _AssertRaisesContext:
    '''
    Fungsi ini ditujukan untuk testing apakah fungsi hitung_penghasilan akan raise TypeError jika menggunakan function dengan
    data type yang salah
    '''
    with self.assertRaises(TypeError):
      return hitung_penghasilan(310239, "presiden")

  def test_total_pengeluaran(self):
    '''
    Fungsi ini ditujukan untuk testing apakah fungsi total_pengeluaran berjalan sesuai ekspektasi
    '''
    karyawan_data = {
      "manager": 50,
      "staff": 50,
      "kasir": 50
    }
    self.assertEqual(total_pengeluaran(karyawan_data), 19600000)

  def test_total_pengeluaran_jenis_pekerjaan_tidak_ada(self):
    '''
    Fungsi ini ditujukan untuk testing apakah fungsi total_pengeluaran akan me-return nilai tetap sesuai ekspektasi,
    meskipun ada data position yang tidak ada di dalam list (data position yang tidak ada di dalam key dictionary
    akan dianggap 0)
    '''
    karyawan_data = {
      "pilot": 50,
      "presiden": 50,
      "astronaut": 50,
      "manager": 45,
      "staff": 35,
      "kasir": 50
    }
    self.assertEqual(total_pengeluaran(karyawan_data), 19300000)

  def test_total_pengeluaran_data_type_salah(self):
    '''
    Fungsi ini ditujukan untuk testing apakah fungsi total_pengeluaran akan me-return nilai 0, jika input dictionary
    diisi dengan tipe data yang salah
    '''
    karyawan_data = {
      "manager": "Joe Biden",
      "staff": "Donald Trump",
      "kasir": "Neil Armstrong",
      "pilot": {
        "nama": "Andreas",
        "umur": 50
      }
    }
    self.assertEqual(total_pengeluaran(karyawan_data), 0)

unittest.main(argv=[''], exit=False)
```
