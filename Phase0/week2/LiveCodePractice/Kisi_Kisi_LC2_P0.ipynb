#####2. Apa saja nama label unique dari tipe sepeda?

#####3. Berapa banyak total peminjaman sepeda pada tahun 2016 dan 2017?

````python
# working area
# rent['year'] = rent['start_time'].dt.year
# print(rent.head())
# rent[(rent["year"] == 2016) | (rent["year"] == 2017)].groupby("year")["trip_id"].count()
the_query3 = """SELECT COUNT(trip_id) AS total_trip, EXTRACT(YEAR FROM start_time) AS year FROM `ftds-hacktiv8-project.practice.bike_rent` GROUP BY year HAVING year = 2017 OR year = 2016""";
quer3 = client.query(the_query3).to_dataframe()
print(quer3)
````

#####4. Dimana station yang paling banyak dikunjungi sebagai titik keberangkatan pada tahun 2016?

````python
# working area
# rent[rent["year"] == 2016].groupby("start_station_name")["trip_id"].count().sort_values(ascending=False)
the_query4 = """SELECT start_station_name, COUNT(trip_id) AS total_station,
EXTRACT(YEAR FROM start_time) AS year FROM `ftds-hacktiv8-project.practice.bike_rent`
WHERE EXTRACT(YEAR FROM start_time) = 2016
GROUP BY start_station_name, year
ORDER BY total_station DESC
""";
quer4 = client.query(the_query4).to_dataframe()
print(quer4)
````

#####5. Berapa nilai minimum, maximum, dan rata-rata dari durasi perjalanan pada keseluruhan data?

````python
# working area
# rent["duration_minutes"].describe()
the_query5 = """SELECT MIN(duration_minutes) AS min_duration, MAX(duration_minutes) AS max_duration, AVG(duration_minutes) AS average_duration
FROM `ftds-hacktiv8-project.practice.bike_rent`
""";
quer5 = client.query(the_query5).to_dataframe()
print(quer5)
````

#####6. Berapa rata-rata durasi perjalanan dari subscriber_type `Student Membership` dan `Local31`?

````python
# working area
# rent.groupby("subscriber_type")["duration_minutes"].mean()
the_query6 = """SELECT  subscriber_type, AVG(duration_minutes) AS average_duration
FROM `ftds-hacktiv8-project.practice.bike_rent`
GROUP BY subscriber_type
HAVING subscriber_type = 'U.T. Student Membership' OR subscriber_type = 'Local30'
""";
quer6 = client.query(the_query6).to_dataframe()
print(quer6)
````

#####7. Tampilkan entry data yang memiliki `duration_time` diatas 4.25 jam. (*Gunakan subquery untuk menjawab pertanyaan ini dengan meng-convert duration_time dari menit ke hour*).

````python
# working area
the_query7 = """
SELECT * FROM `ftds-hacktiv8-project.practice.bike_rent`
WHERE trip_id IN (SELECT trip_id
FROM `ftds-hacktiv8-project.practice.bike_rent`
WHERE ROUND(duration_minutes / 60, 2) > 4.25)
""";
quer7 = client.query(the_query7).to_dataframe()
print(quer7)
````

#####2. Buat kolom baru bernama `quarter`.
  
  Lakukan ekstraksi waktu (quarter dalam tahun) berdasarkan kolom `start_time` dengan ketentuan berikut:

<div align="center">

  |Keterangan Waktu|Bulan|
  |--|--|
  |Q1|Januari, Februari, Maret|
  |Q2|April, Mei, Juni|
  |Q3|Juli, Agustus, September|
  |Q4|Oktober, November, Desember|

</div>

>**Notes:**
>- Dalam menjawab soal Pandas no. 2, **tidak boleh** menggunakan ekstraksi waktu `Series.dt.quarter`.
- Lakukan `looping` dan `conditional` dengan memanfaatkan ekstraksi bulan `Series.dt.month` terhadap kolom `start_time`, pastikan tipe datanya ialah `datetime` agar dapat dilakukan ekstraksi.
- Simpan hasil looping ke dalam suatu variabel, dan buat kolom baru berisikan variabel sebelumnya.

````python
# working area
df1["quarter"] = df1["start_time"].dt.month
for i, row in df1.iterrows():
  if row["start_time"].month >= 1 and row["start_time"].month <= 3:
    df1.at[i, 'quarter'] = "Q1"
  elif row["start_time"].month >= 4 and row["start_time"].month <= 6:
    df1.at[i, 'quarter'] = "Q2"
  elif row["start_time"].month >= 7 and row["start_time"].month <= 9:
    df1.at[i, 'quarter'] = "Q3"
  else:
    df1.at[i, 'quarter'] = "Q4"
df1
````

#####4. Berapa rata-rata durasi perjalanan dari tipe pelanggan `Walk Up` dengan tipe sepeda `classic` pada bulan `Maret` dari data yang diambil?
*(Jika tidak ada entry yang memenuhi kondisi ini tidak apa, tuliskan kode pandasnya dengan lengkap saja)*
